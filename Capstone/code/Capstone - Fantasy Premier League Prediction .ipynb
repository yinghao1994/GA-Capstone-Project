## Executive Summary

Fantasy Premier League (FPL) is a fantasy sport management game; the aim being to score as many fantasy points as possible at the end of the football season. The results will be tabulated and the top scoring players will stand to win several attractive prizes. 

A season consists of 38 gameweeks. At the beginning, fantasy players are given a virtual currency budget of $100 million. During each gameweek, players will have to pick a total of 15 footballers from the pool that is available in the Premier League. The draft consists of 2 Goalkeepers, 5 Defenders, 5 Midfielders and 3 Forwards. At any given gameweek, only 11 are allowed to be fielded (1 GK, 3-5 Defenders, 3-5 Midfielders, 1-3 Forwards) and 4 benched. 

After the matches are played for each week, points are awarded based on how the statistics they have racked up in the match. Points can be awarded for minutes played, goals, assists, influence on the pitch etc. The default strategy of FPL is to pick good footballer that have the potential to rack up as many points as possible. However, it is also important that the fantasy players choose their footballers wisely as they have a limited budget. 

## Problem Statement 

The aim of this project would be to find out if it is possible to predict players who would perform well in the Fantasy Premier League from a machine-learning algorithm. We would want to build a machine-learning model to predict the a player's total points for the season based on the footballers stats.

## Data Cleaning

```python
# Fuction that merges all gameweek data within a season 

def merge_gw(season):
    
    df = pd.read_csv('datasets/' + season + '/gw1.csv', engine = 'python')
    df['GW'] = 1
    
    for i in range(2,39):
        
        df2 = pd.read_csv('datasets/' + season + '/gw' + str(i) + '.csv' , engine = 'python' )
        df2['GW'] = i  # Labelling data that corresponds to the appropriate gameweek
        df = pd.concat([df,df2], axis = 0)
    
    return df
```

For each FPL season, players come and go during the transfer window period at each season break. Thus, in order to accurately cross-validate the train data (Seasons 2016-2018) and test data (2018 - 2019), we need to intersect the data such that the same players are present in the train and test dataset.

```python
# Function is to intersect two lists together 

def intersect_list(lst1, lst2):
    new_lst = []
    for i in lst1:
        if i in lst2:
            new_lst.append(i)
    
    return new_lst
```

```python
# Creating a dataframe of player names, so that I can merge and drop the redundant list of both train and test datasets 
final_playerlist = intersect_list(player_1618_namelist,player_1819_namelist)
final_playerdf = pd.DataFrame(final_playerlist, columns = ['full_name'])
final_playerdf
```

## Exploratory Data Analysis 

Given we have a completed dataset for all 38 weeks in a season, a quick way to conduct EDA would be to use the data cumulated at the end of the season (i.e. Gameweek 38). We will be determining which features are more important in obtaining the best possible score.

```python
plt.figure(figsize=(8,5))
plt.scatter(player_1618_final['ict_index'],player_1618_final['total_points']);
plt.title('Player Total Points v.s ICT Index');
plt.ylabel('Total Points');
plt.xlabel('ICT Index');
```

The above scatterplot represents the total points of each player versus their ICT index. ICT is an acronym for influence, creativity and threat, which are indexes formulated by the fantasy premier league game designers given their recent match performances. These indexes are calculated by a multitude of features, which includes goals, assist, bonus points, and even to details such as passes completed, tackles made etc. This formula is kept privy by the FPL gamemasters. 

From the scatterplot, it can observe that there is a high correlation between total points and ICT index. This shows that the ICT index developed by FPL is a relatively good gauge of a player's skill level. 

```python
plt.figure(figsize=(20,6))

plt.subplot(1,2,1)
plt.scatter(player_1618_final['goals_scored'],player_1618_final['total_points']);
plt.title('Player Total Points v.s Goals Scored');
plt.ylabel('Total Points');
plt.xlabel('Goals Scored');

plt.subplot(1,2,2)
plt.scatter(player_1618_final['assists'],player_1618_final['total_points']);
plt.title('Player Total Points v.s Assist');
plt.ylabel('Total Points');
plt.xlabel('Assist');
```

Both scatterplots look very similar. It can be observed that the scatter points cluster away from the bottom right of the graph. This is so as goals and assists are awarded FPL points; it is impossible to have a lot of goals/assists but with no points. What's interesting to note is that points are still awarded for footballers who have not scored or assisted. This means that there are methods to gain points besides scoring goals or creating assists.   

```python
plt.figure(figsize=(20,4))

plt.subplot(1,3,1)
plt.scatter(player_1618_final['influence'],player_1618_final['total_points']);
plt.title('Player Total Points v.s Influence Index');
plt.ylabel('Total Points');
plt.xlabel('Influence Index');

plt.subplot(1,3,2)
plt.scatter(player_1618_final['creativity'],player_1618_final['total_points']);
plt.title('Player Total Points v.s Creativity Index');
plt.ylabel('Total Points');
plt.xlabel('Creativity Index');

plt.subplot(1,3,3)
plt.scatter(player_1618_final['threat'],player_1618_final['total_points']);
plt.title('Player Total Points v.s Threat Index');
plt.ylabel('Total Points');
plt.xlabel('Threat Index');
```

The ICT Index is broken down into its individual portions: Influence, Creativity and Threat Index. 

Threat index is a gauge on how dangerous a player is against the opposition, his ability to shoot on target and score goals. Creativity Index gauges how players can create chances for other players to score goals. If we juxtapose the scatter pattern of these two graphs against goals and assists made, we notice that the scatter is very similar. Basically, your threat and creativity index is heavily reliant on goals and assist made respectively. 
Influence index is a gauge on how a player is able to impact the match as a whole, the amount of influence he has on the game. This is governed by a number of overall statistical measures which are privy to FPL gamemasters. Unlike Threat or Creativity, influence indexed is calculated using a slew of statistics, gauging the overall impact a footballer has in the game.

### Linear Regression

```python
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNet, ElasticNetCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.tree import DecisionTreeClassifier
```

```python
player_1618_final.columns
```

```python
# Features selected based on EDA for linear regression
features = ['goals_scored', 'assists', 'minutes',
       'goals_conceded', 'creativity', 'influence', 'threat', 'bonus', 'bps',
       'ict_index', 'clean_sheets', 'red_cards', 'yellow_cards',
       'selected_by_percent']
```

```python
X_train_linreg = player_1618_final[features]
y_train_linreg = player_1618_final['total_points']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train_linreg, y_train_linreg, test_size=0.2, random_state=42)
lr = LinearRegression()
lr_model = lr.fit(X_train,y_train)

#Creating variables for statistical metrics 
coef  = lr.coef_
intercept = lr.intercept_
x_val_score = cross_val_score(lr, X_train,y_train, cv =5)
x_val_score_mean = np.mean(x_val_score)
lr_score_train = lr_model.score(X_train,y_train)
lr_score_test = lr_model.score(X_test,y_test)
lr_predict = lr_model.predict(X_test)
RMSE = np.sqrt(mean_squared_error(lr_predict, y_test))

print('LR Model Coefficients: ' + str(coef))
print('LR Model Intercept: ' + str(intercept))
print('LR Model Cross Validation Scores: ' + str(x_val_score))
print('LR Model Cross Validation Mean Score: ' + str(x_val_score_mean))
print('LR Model R-Squared Train Score: ' + str(lr_score_train))
print('LR Model R-Squared Test Score: ' + str(lr_score_test))
print('LR Model RMSE: ' + str(RMSE))

# Linear Regression Scatterplot
plt.figure(figsize=(18,12));
plt.scatter(lr_predict,y_test);
plt.xlabel('Predicted Total Points', fontsize = 16);
plt.ylabel('True Total Points', fontsize = 16);
plt.title('Linear Regression Model Scatterplot on Total Points of FPL Footballers', fontsize = 30);
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], c='red');
```

```python
# Linear Regression Model Intialised used to predict FPL test data (season 18/19)

lr_predict = lr_model.predict(player_1819_final[features])
RMSE = np.sqrt(mean_squared_error(lr_predict, player_1819_final['total_points']))

lr_score_test_final = lr_model.score(player_1819_final[features],player_1819_final['total_points'])

print('LR Model R-Squared Test Score: ' + str(lr_score_test_final))
print('LR Model RMSE: ' + str(RMSE))

# Generating Model Predictions for the test dataset of season 18/19

fpl_lr = pd.DataFrame({'Player Name 1618': player_1618_final['full_name'],  'Predicted Points': lr_predict,
                       'Actual Points':player_1819_final['total_points']})

# you could use any filename. We choose submission here
fpl_lr.to_csv('datasets/fpl_lr.csv', index=False)
```

### Conclusion

From the above models, we notice certain commonalities between key variables in models that have displayed a high sensitivity in prices, which are:
- Minutes
- Bonus Points System
- Goals Scored 
- Assists 
- Clean Sheet (Not letting in any goals)

To sum up, linear regression model has shown to be on par with models such as lasso and ridge regression models. Features such as minutes, goals and assists are still reliable predictors on how a player would do in future FPL Seasons. An important reason for this would be that most of the features are directly correlated to points obtained by a certain footballer. For example, if a player plays the full 90 minutes of the game, he would have more chances to score goals, create chances, getting clean sheets. Thus, the minute variable is the variable which has the largest coefficient. 


### Limitations

A major weakness of the model is that we do not take into account how the player performs over the season. 
There will be times where player form dips and we are not able to pick up on that. 
If given the appropriate data to model after subsequent gameweeks, we could use time-series modelling to account for the footballers playing matches week-in week-out. 

The models also do not take into account whether the player would be injured, or when they would have a dip in form and thus the manager would not select them. A player might do well for the first half of the season, but may be injured and there is no way of predicting when that will happen.
